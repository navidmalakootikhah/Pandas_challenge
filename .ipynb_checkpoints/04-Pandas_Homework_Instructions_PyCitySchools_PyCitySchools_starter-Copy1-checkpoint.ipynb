```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

```python
school_count=school_data["school_name"].count()
student_count=student_data["student_name"].count()
total_budget=school_data["budget"].sum()

average_math=student_data["math_score"].mean()
average_reading=student_data["reading_score"].mean()
overall_passing_rate=(average_math+average_reading)/2
math_passing_rate=(len(student_data.loc[student_data["math_score"]>=70,["math_score"]])/len(student_data["math_score"]))*100
reading_passing_rate=(len(student_data.loc[student_data["reading_score"]>=70,["reading_score"]])/len(student_data["reading_score"]))*100
summary_student=pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math": [math_passing_rate],
    "% Passing Reading": [reading_passing_rate],
    "% Overall Passing Rate": [overall_passing_rate]})

summary_student["Total Budget"] = summary_student["Total Budget"].map("${:,}".format)


summary_student.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_counts = school_data_complete["school_name"].value_counts()
school_group=school_data_complete.groupby('school_name')
student_count=school_data_complete["school_name"].value_counts()
total_school_budget=school_group["budget"].sum()/school_counts


per_student_budget=total_school_budget/student_count
average_math=school_group["math_score"].sum()/student_count
average_reading=school_group["reading_score"].sum()/student_count


#calculate the passing rates for math
math_greater_70=(school_data_complete.loc[student_data["math_score"]>=70,["school_name"]])
math_greater_70=math_greater_70["school_name"].value_counts()
math_pass=(math_greater_70/student_count)*100
math_pass

#calculate the passing rates for reading
reading_greater_70=(school_data_complete.loc[student_data["reading_score"]>=70,["school_name"]])
reading_greater_70=reading_greater_70["school_name"].value_counts()
reading_pass=(reading_greater_70/student_count)*100
reading_pass

#calculate the passing rates for overall
overals_pass=((math_pass+reading_pass)/2)
overals_pass

#create a data frame
school_type=school_group["type"].unique().astype(str)

#school_type = school_type.rename(columns={"type": "count","school_name"})
#del school_type['count']


summary_school=pd.DataFrame({"District":school_type,"Total Student":student_count,"Total School Budget":total_school_budget,"Per Student Budget":per_student_budget,"Average Math Score":average_math,"Average Reading Score":average_reading,"% Passing Math":math_pass,"% Reading Pass":reading_pass,"% Overall Passing Rate":overals_pass})
summary_school["Total School Budget"] = summary_school["Total School Budget"].map("${:,}".format)

summary_school = summary_school.sort_values(["% Overall Passing Rate"], ascending=True)

summary_school.head()

```
